```python

fig=plt.figure()
ax=plt.axes()
x= np.linspace(0,10,1000)
print(x)
ax.plot(x,np.cos(x));
ax.plot(x,np.sin(x));
ax.set_title('simple plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label');
```

```python
fig=plt.figure()
ax=plt.axes()
x= np.linspace(0,10,1000)
print(x)
ax.plot(x,np.cos(x), label='cos');
ax.plot(x,np.sin(x), label='sin');
ax.set_title('multiple lines')
ax.set_xlabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
x= np.linspace(0,10,1000)
print(x)
ax.plot(x,np.cos(x), label='cos', color='black');
ax.plot(x,np.sin(x), label='sin', color='pink');
ax.set_title('multiple lines')
ax.set_xlabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
x= np.linspace(0,10,1000)
print(x)
ax.plot(x,np.cos(x), label='cos', color='pink', linestyle='dashed');
ax.plot(x,np.sin(x), label='sin', color='orange', linestyle='dashdot');
ax.set_title('multiple lines')
ax.set_xlabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```
